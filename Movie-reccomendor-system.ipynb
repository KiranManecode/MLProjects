```python
import ast
```

```python
def convert(text):
  L=[]
  for i in ast.literal_eval(text):
    L.append(i['name'])
  return L

```

```python
movies.dropna(inplace=True)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies
```

```python
movies.isnull().sum
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter += 1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
newf=movies[['movie_id','title','tags']]
```

```python
newf
```

```python
newf['tags'] = newf['tags'].apply(lambda x: " ".join(x))
```

```python
newf.head()
```

```python
newf['tags'][0]
```

```python
newf['tags']=newf['tags'].apply(lambda x:x.lower())
```

```python
newf
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(newf['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
newf[newf['title']=='Avatar'].index[0]
```

```python
def recommend(movie):
  movie_index=newf[newf['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(newf.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
